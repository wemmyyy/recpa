```python
#numero impar ou par

numero = int(input("digite um numero: "))

if numero % 2 == 0:
  print("numero par.")

else:
  print("numero impar.")
```

```python
#dois numeros

numero1 = float(input("digite um numero: "))
numero2 = float(input("digite um numero: "))

if numero1 > numero2:
  print(f"o numero {numero1} é maior que o numero {numero2}")

elif numero2 > numero1:
  print(f"o numero {numero2} é maior que o numero {numero1}")

elif numero1 == numero2:
  print(f"o numero {numero1} é igual ao numero {numero2}")
```

```python
#nota

nota = float(input("digite sua nota: "))

if nota >= 7:
  print("aprovado.")

elif nota >= 5 and nota < 7:
  print("recuperação.")

else:
  print("reprovado.")
```

```python
#idade

idade = int(input("digite sua idade: "))

if idade <= 12:
  print("criança.")

elif idade >= 13 and <= 17:
  print("adolescente.")

elif idade >= 18 and <= 59:
  print("adulto.")

else:
  print("idoso.")
```

```python
#senha

senha = str(input("digite sua senha: "))

if senha == "Senha123":
  print("senha correta.")

else:
  print("senha incorreta.")
```

```python
#negativo, positivo ou zero

numero = float(input("digite um numero: "))

if numero > 0:
  print("numero positivo.")

elif numero < 0:
  print("numero negativo.")

else:
  print("zero.")
```

```python
#calculadora

numero1 = float(input("digite um numero: "))
operacao = input("Digite a operação (+, -, *, /): ")
numero2 = float(input("digite um numero: "))

mais = numero1 + numero2
menos = numero1 - numero2
vezes = numero1 * numero2
divisao = numero1 / numero2

if operacao == '+':
  print(f"resultado igual a: {mais}")

elif  operacao == '-':
  print(f"resultado igual a: {menos}")

elif  operacao == '*':
  print(f"resultado igual a: {vezes}")

elif  operacao == '/':
  print(f"resultado igual a: {divisao}")
```

```python
#Dia da semana

semana = int(input("digite o dia da semana: "))

if semana == 1:
  print("segunda")

elif semana == 2:
  print("terça")

elif semana == 3:
  print("quarta")

elif semana == 4:
  print("quinta")

elif semana == 5:
  print("sexta")

elif semana == 6:
  print("sabado")

elif semana == 7:
  print("domingo")
```

```python
#Ano bissexto

ano = int(input("Digite o ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"{ano} é bissexto.")
else:
    print(f"{ano} não é bissexto.")
```

```python
#Desconto em compras

valor = float(input("digite o valor da compra: R$"))

if valor >= 100:
  print("desconto de 10%")

elif valor >= 50:
  print("desconto de 5%")

else:
  print("sem desconto")
```
